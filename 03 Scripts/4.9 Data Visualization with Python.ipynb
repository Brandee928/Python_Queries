```python
#Pandas & Visual Library
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
path = r'/Users/brandeefairnot/Library/CloudStorage/OneDrive-Personal/Documents/Instacart Basket Analysis/'
```

```python
#import file
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'df_merged_new_4.8.pkl'))
```

## creating bar chart

```python
#order of largest to smallest
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
#sorting bar charts in chronological order
ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
#adding color
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
#how to 'call' a graph
bar = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
#how to export a graph 

bar.figure.savefig(os.path.join(path, '04 Analysis', 'bar_orders_dow.png'))
```

## creating histograms & scatterplots

```python
# creating histogram
ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# creating scatterplot
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
```

```python
# marking outliers NA
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
#check that removing outliers worked
ords_prods_merge['prices'].max()
```

```python
#run histogram functino again
hist = ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
#saving histogram 
hist.figure.savefig(os.path.join(path, '04 Analysis', 'hist_orders_dow.png'))
```
